```python
import pywt
# Import python library for wavelet decomposition
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
pressure_data = np.linspace(0,10,1000)
```

```python
random_data=np.random.normal(scale=0.5,size=1000)
#pressure_data+=
```

```python
plt.scatter(range(len(pressure_data)),pressure_data)
```

```python
plt.scatter(range(len(random_data)),random_data)
```

```python
pressure_data +=random_data
```

```python
plt.scatter(range(len(pressure_data)),pressure_data)
```

```python
# Define the wavelet
wavelet='db4'
level=4
```

```python
coef = pywt.wavedec(pressure_data,wavelet,level=level)
```

```python
coef
```

```python
np.shape(coef)
```

```python
AC = coef[0]
DC1 = coef[1]
DC2 = coef[2]
DC3 = coef[3]
DC4 = coef[4]
```

```python
def cal_engergy(data):
    return np.sum(np.square(data))
```

```python
AC_Energy = cal_engergy(AC)
DC1_Energy = cal_engergy(DC1)
DC2_Energy = cal_engergy(DC2)
DC3_Energy = cal_engergy(DC3)
DC4_Energy = cal_engergy(DC4)
```

```python
print(AC_Energy,DC1_Energy,DC2_Energy,DC3_Energy,DC4_Energy)
```

```python
total_energy = [np.sum(np.square(cf)) for cf in coef]
```

```python
total_energy=np.sum(total_energy)
```
